**P.S :** Vous obtiendrez probablement des erreurs ci-dessous en raison d'une limitation dans l'utilisation de l'API Spotify.

```python
# Initialiser les listes pour stocker les donn√©es
track_names = []
artist_names = []
album_names = []
release_dates = []
durations_ms = []
popularity_scores = []
track_ids = []

# Requ√™te pour obtenir les titres des Beatles
offset = 0
limit = 50  # Maximum autoris√© par Spotify pour une requ√™te
while True:
    response = requests.get(f"https://api.spotify.com/v1/search?q=artist:The%20Beatles&type=track&limit={limit}&offset={offset}",
                            headers=headers)
    try:
      tracks_data = response.json()['tracks']['items']

      if not tracks_data:
          break  # Sortie de la boucle si aucun titre n'est retourn√©

      for track in tracks_data:
          track_names.append(track['name'])
          artist_names.append(', '.join(artist['name'] for artist in track['artists']))  # Les Beatles peuvent avoir des artistes invit√©s
          album_names.append(track['album']['name'])
          release_dates.append(track['album']['release_date'])
          durations_ms.append(track['duration_ms'])
          popularity_scores.append(track['popularity'])
          track_ids.append(track['id'])

      offset += limit  # Pr√©parer l'offset pour la prochaine requ√™te
    except:
      print(f"Une erreur est peut-√™tre survenue avec la r√©ponse {response}")
```

```python
beatles_tracks_df = pd.DataFrame({
    'Track ID': track_ids,
    'Track Name': track_names,
    'Artists': artist_names,
    'Album Name': album_names,
    'Release Date': release_dates,
    'Duration (ms)': durations_ms,
    'Popularity': popularity_scores
})

# Convertir la dur√©e de millisecondes en minutes:secondes pour plus de lisibilit√©
beatles_tracks_df['Duration (ms)'] = beatles_tracks_df['Duration (ms)'].apply(lambda x: f"{x // 60000}:{(x % 60000) // 1000:02d}")

beatles_tracks_df.head()
```
